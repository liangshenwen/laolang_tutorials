# 有理数的基本运算（加，减，乘，除，乘方）

```python
# 加法
print('1 + 2 = ', 1 + 2)  # 整数
from fractions import Fraction
print('1/2 + 1/3 = ', Fraction(1,2) + Fraction(1,3)) # 分数  
```

```python
# 减法
print('1 - 2 = ', 1 - 2)  # 整数
from fractions import Fraction
print('1/2 - 1/3 = ', Fraction(1,2) - Fraction(1,3)) # 分数 
```

```python
# 乘法
print('1 X 2 = ', 1*2)  # 整数
from fractions import Fraction
print('1/2 X 1/3 = ', Fraction(1,2) * Fraction(1,3)) # 分数 
```

```python
# 除法
print(f'1 {chr(247)} 2 = ', 1/2)  # 整数
from fractions import Fraction
print(f'1/2 {chr(247)} 1/3 = ', Fraction(1,2) / Fraction(1,3)) # 分数 
```

什么是乘方?  


```python
# 乘方
import math
print(f'2^3 = ', 2**3)  # 整数
print(f'2^3 = ', math.pow(2,3))  # 整数
from fractions import Fraction
print(f'(1/2)^3 = ', Fraction(1,2) ** 3) # 分数
print(f'(1/2)^3 = ', math.pow(Fraction(1,2), 3)) # 分数
```

```python
# 混合运算
print(f'(1 + 2^3) - 3/6 + 9 * 3 = ', (1 + 2**3) - 3/6 + 9 * 3)
```

# 一元一次方程
一元一次方程指只含有一个未知数、未知数的最高次数为1且两边都为整式的等式。一元一次方程只有一个根。一元一次方程可以解决绝大多数的工程问题、行程问题、分配问题、盈亏问题、积分表问题、电话计费问题、数字问题。

一元一次方程的标准形式是：ax+b=0 (其中x是未知数，a,b是已知数，且a≠0)



```python
from sympy import *

x = symbols('x')
z = solve(x/70 - x/60 + 1, x)
print(round(z[0]),'km')
```

```python
from sympy import *

x = symbols('x')
z = solve(0.3*x + 0.6*(20-x) - 9, x)
# round(number[, ndigits]): 
# Return number rounded to ndigits precision after the decimal point. 
#    If ndigits is omitted or is None, it returns the nearest integer to its input.

print(round(z[0]),'支甲种铅笔')
print(20-round(z[0]),'支乙种铅笔')
```

```python
from sympy import *

x = symbols('x')
z = solve(0.52*x - (1-0.52)*x - 80, x)

print('这个学校有', round(z[0]),'个学生')
```

```python
from sympy import symbols, solve
from fractions import Fraction

# 使用float来进行运算，会导致数值误差
x = symbols('x')
z = solve(x/0.7 - (0.17-0.2*x)/0.03 - 1, x)
print('x =',  z[0])
print(type(z[0]))
print(Fraction(str(z[0])))
print(14/17)
```

```python
# sympy 如何使用分数(有理数)进行运算?
from sympy import Integer,symbols,solve
from fractions import Fraction

x = symbols('x')
z = solve(x/(Integer(7)/Integer(10)) - (Integer(17)/Integer(100) - (Integer(2)/Integer(10))*x)/(Integer(3)/Integer(100)) - 1, x)
print('x =',  z[0])
print(type(z[0]))
```

```python
# sympy 如何使用分数(有理数)进行运算
from sympy import Rational,symbols,solve
from fractions import Fraction

x = symbols('x')
z = solve(x/Rational(7,10) - (Rational(17,100) - Rational(2,10) * x) /Rational(3,100) - 1, x)
print('x =',  z[0])
print(type(z[0]))
```

## 五角星
标准的五角星怎么画    
一、工具：白纸，圆规，直尺，铅笔   

二、步骤： 
1.在白纸上，以任意一点为圆心，以任意长为半经画圆O   
  

2.在圆中画两条互相垂直的圆的直经AB和CD.  
  

3.取线段OB的中点E,连接CE.  
  

4.以点E为圆心，以CE长为半经画圆弧，交线段OA于点F.  
  

5.，连接CF,以点C为圆心，以CF长为半经在圆O上依次截取相等的圆弧   
  

6.连接CM、CH、GN、GM、NH,就得到正五角星   
  

7.沿着正五角星的形状，剪下正五角星   
  


五角星的角度数为36度
