```python
import pandas as pd
import numpy as np
import baostock as bs

pd.set_option('display.max_rows', 20)

encoding = 'UTF-8'

sz50_stocks_csv_file = 'sz50_stocks.csv'
hs300_stocks_csv_file = 'hs300_stocks.csv'
zz500_stocks_csv_file = 'zz500_stocks.csv'
```

# 上证50成分股

```python
import baostock as bs
import pandas as pd

# 登陆系统
lg = bs.login()
# 显示登陆返回信息
# print('login respond error_code:'+lg.error_code)
# print('login respond  error_msg:'+lg.error_msg)

# 获取上证50成分股
rs = bs.query_sz50_stocks()
# print('query_sz50 error_code:'+rs.error_code)
# print('query_sz50  error_msg:'+rs.error_msg)

# 打印结果集
sz50_stocks = []
while (rs.error_code == '0') & rs.next():
    # 获取一条记录，将记录合并在一起
    sz50_stocks.append(rs.get_row_data())
result = pd.DataFrame(sz50_stocks, columns=rs.fields)
# 结果集输出到csv文件
result.to_csv(sz50_stocks_csv_file, encoding=encoding, index=False)

# 登出系统
bs.logout()

result
```

```python
import baostock as bs
import pandas as pd

def download_stock_prices(stock_codes, csv_file):
    lg = bs.login()
    stock_prices = []
    for stock_code in stock_codes:
        #### 获取沪深A股历史K线数据 ####
        rs = bs.query_history_k_data_plus(stock_code, 
                                          "date,code,open,high,low,close,preclose,volume,amount,adjustflag, turn,tradestatus,pctChg,isST",
                                          frequency="d", 
                                          start_date="1900-01-01",
                                          adjustflag="3")
        while (rs.error_code == '0') & rs.next():
            # 获取一条记录，将记录合并在一起
            stock_prices.append(rs.get_row_data())
        
        print("stock code", stock_code , "is done.")
    
    stock_prices_df = pd.DataFrame(stock_prices, columns=rs.fields)
    #### 结果集输出到csv文件 ####   
    stock_prices_df.to_csv(csv_file, index=False)

    #### 登出系统 ####
    bs.logout()
        
```

# 画某个股票的日K线图

在使用 Python 进行金融数据分析时，绘制 K线图 是很常见的需求。 

但如果使用 matplotlib 从 0 开始绘制，一步一步添加日线、均线、MACD、成交量等指标时，则会显得十分麻烦，且代码很难复用。 

幸运的是在 matplotlib 中提供接口（matplotlib.finance）直接绘制K线，现在 matplotlib.finance 已经独立成库 mplfinance，更方便的让我们使用。 

```python
! python -m pip install matplotlib mplfinance > pip_install.log
```

```python
stock_prices_1_df = zz500_prices_df.loc[zz500_prices_df['code']=='sh.600000', ['date','open', 'high', 'low', 'close', 'volume']]
stock_prices_1_df.rename(str.capitalize, axis='columns', inplace=True)
stock_prices_1_df.set_index('Date', inplace=True)
stock_prices_1_df
```

```python
import mplfinance as mpf

stock_prices_1_df = stock_prices_1_df.tail(30)
```

## 可生成简易价格走势图

```python
mpf.plot(stock_prices_1_df, type='line')
```

## 添加移动均线

通过设置 mav 参数可以添加对应的移动均线，例如添加5日、10日、30日移动均线

```python
mpf.plot(stock_prices_1_df, type='line',mav=(5,10,30))
```

需要注意的是，这里的x日移动均线并不是通过我们数据的时间索引计算而来，仅是移动x个索引位置而来，由于我们数据时间精度为1天，所以恰好是对应的x天移动均线。

如果索引是分钟级数据，那么设置 mav 得到的就是x分钟均线！

```python
## 添加成交量
通过设置 volume 参数，可以进一步添加成交量
```

```python
mpf.plot(stock_prices_1_df, type='line',mav=(5,10,30), volume=True)
```

## 制作蜡烛图
通过设置 type 参数可以绘制更专业的蜡烛图，由于上面的数据时间维度过长，绘制蜡烛图会导致很难看清细节。

所以下面绘制最后 40 个交易日的蜡烛图，并添加成交量与 3、6、9 日均线

```python
mpf.plot(stock_prices_1_df, type='candle',mav=(3,6,9), volume=True)
```

## 展示非交易时间
上面的图是连续的，但交易日并不是天天都是，每天也有指定时间，通过设置 show_nontrading 参数，可以按照交易时间绘制，将非交易时间添加为空白

```python
mpf.plot(stock_prices_1_df,type='candle',mav = (3),show_nontrading=True, volume=True)
```
