```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

zh_font = fm.FontProperties(fname="C:\\Windows\\Fonts\\simsun.ttc", size=18)
```

# 三角函数线


正弦（Sine）：sinA = ∠A的对边与斜边的比  
余弦（Cosine） ：cosA = ∠A的邻边与斜边的比  
正切（Tangent）：tanA = ∠A的对边与邻边的比  
sin2A + cos2A =  1

  

正割sec是余弦的倒数：sec α = 1/cos α  
余割csc是正弦的倒数：csc α = 1/sin α  



# 正弦函数


```python
np.sin(np.array((0., 30., 45., 60., 90.)) * np.pi / 180. )
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 401)
plt.plot(x, np.sin(x))
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.title('正弦函数',fontproperties=zh_font)
plt.grid()
plt.show()
```

# 余弦函数


```python
x = np.linspace(-2*np.pi, 2*np.pi, 401)
plt.plot(x, np.cos(x))
plt.xlabel('Angle [rad]')
plt.ylabel('cos(x)')
plt.title('余弦函数',fontproperties=zh_font)
plt.grid()
plt.show()
```

# 正切函数
这里值得注意的是正切函数是以PI为周期，以及关于原点对称，奇函数。  
  


```python
x = np.arange(-0.49*np.pi,0.49*np.pi,0.01)
plt.plot(x, np.tan(x))
plt.xlabel('Angle [rad]')
plt.ylabel('tan(x)')
plt.title('正切函数',fontproperties=zh_font)
plt.grid()
plt.show()
```

# 余切函数，余割函数，正割函数


```python
x = np.arange(0.01*np.pi,0.99*np.pi,0.01)
y = np.tan(x)
y = 1/y
plt.plot(x, y)
plt.xlabel('Angle [rad]')
plt.ylabel('cot(x)')
plt.title('余切函数',fontproperties=zh_font)
plt.grid()
plt.show()
```

```python
x = np.arange(-0.49*np.pi, 0.49*np.pi, 0.001)
plt.plot(x, 1/np.cos(x))
plt.xlabel('Angle [rad]')
plt.ylabel('sec(x)')
plt.title('正割函数',fontproperties=zh_font)
plt.grid()
plt.show()
```

```python
x = np.arange(0.01*np.pi, 0.99*np.pi, 0.001)
plt.plot(x, 1/np.sin(x))
plt.xlabel('Angle [rad]')
plt.ylabel('csc(x)')
plt.title('余割函数',fontproperties=zh_font)
plt.grid()
plt.show()
```
