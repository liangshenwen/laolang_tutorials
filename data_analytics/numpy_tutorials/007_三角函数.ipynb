```python
x = np.linspace(-2*np.pi, 2*np.pi, 401)
plt.plot(x, np.sin(x))
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.title('',fontproperties=zh_font)
plt.grid()
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 401)
plt.plot(x, np.cos(x))
plt.xlabel('Angle [rad]')
plt.ylabel('cos(x)')
plt.title('',fontproperties=zh_font)
plt.grid()
plt.show()
```

# 
PI  
  


```python
x = np.arange(-0.49*np.pi,0.49*np.pi,0.01)
plt.plot(x, np.tan(x))
plt.xlabel('Angle [rad]')
plt.ylabel('tan(x)')
plt.title('',fontproperties=zh_font)
plt.grid()
plt.show()
```

# 


```python
x = np.arange(0.01*np.pi,0.99*np.pi,0.01)
y = np.tan(x)
y = 1/y
plt.plot(x, y)
plt.xlabel('Angle [rad]')
plt.ylabel('cot(x)')
plt.title('',fontproperties=zh_font)
plt.grid()
plt.show()
```

```python
x = np.arange(-0.49*np.pi, 0.49*np.pi, 0.001)
plt.plot(x, 1/np.cos(x))
plt.xlabel('Angle [rad]')
plt.ylabel('sec(x)')
plt.title('',fontproperties=zh_font)
plt.grid()
plt.show()
```

```python
x = np.arange(0.01*np.pi, 0.99*np.pi, 0.001)
plt.plot(x, 1/np.sin(x))
plt.xlabel('Angle [rad]')
plt.ylabel('csc(x)')
plt.title('',fontproperties=zh_font)
plt.grid()
plt.show()
```
