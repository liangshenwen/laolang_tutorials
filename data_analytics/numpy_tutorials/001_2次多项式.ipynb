```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
def my_func(x):
    return 3*x**2 + 2*x + 1

x = np.linspace(-2, 2)

y = my_func(x)
zh_font = fm.FontProperties(fname="C:\\Windows\\Fonts\\simsun.ttc", size=18)
plt.title("多项式曲线",fontproperties=zh_font)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(linestyle='--', alpha=0.5)
plt.plot(x, y,'o')
plt.show()
```
