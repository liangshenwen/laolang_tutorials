# 均匀分布随机数

```python
n = 800 # 样本数
x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x, y) # 绘制散列图
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

从上图我们可以看出随机数是均匀分布的。

# 正态分布随机数

```python
n = 800 # 样本数
x = np.random.randn(n)
y = np.random.randn(n)

plt.scatter(x, y) # 绘制散列图
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
