On DataFrame, plot() is a convenience to plot all of the columns with labels

```python
import pandas as pd
import numpy as np
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts
```

```python
ts.plot()
```

```python
ts_cumsum = ts.cumsum()
ts_cumsum
```

```python
ts_cumsum.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list("ABCD"))
df
```

```python
df_cumsum = df.cumsum()
df_cumsum
```

```python
df_cumsum.plot(kind='line')
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=["B", "C"])
df3_cumsum = df3.cumsum()
df3["A"] = pd.Series(list(range(len(df3))))
df3_cumsum["A"] = pd.Series(list(range(len(df3_cumsum))))
df3
```

```python
df3.plot(x="A", y="B")
```

```python
df3_cumsum
```

```python
df3_cumsum.plot(x="A", y="B")
```

```python
df3_cumsum.plot(x="A", y=["B","C"])
```

```python
df3_cumsum.plot(x="A", y=["B","C"], legend=False)
```

```python
df3_cumsum.plot.line(x="A", y=["B","C"], legend=False)
```
