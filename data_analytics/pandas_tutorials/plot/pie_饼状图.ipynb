```python
import pandas as pd
import numpy as np 
import matplotlib as plt
```

```python
series = pd.Series(3 * np.random.rand(4), index=["a", "b", "c", "d"], name="series")
series
```

```python
series.plot.pie(figsize=(6, 6))
```

```python
series.plot.pie(
    labels=["AA", "BB", "CC", "DD"],
    colors=["r", "g", "b", "c"],
    autopct="%.2f",
    fontsize=20,
    figsize=(6, 6),
)
```

```python
df = pd.DataFrame(
    3 * np.random.rand(4, 2), index=["a", "b", "c", "d"], columns=["x", "y"]
)
df
```

```python
df.plot.pie(subplots=True, fontsize=20,figsize=(8, 4))
```
